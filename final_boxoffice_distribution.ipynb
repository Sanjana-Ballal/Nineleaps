```python
import pandas as pd
import matplotlib.pyplot as plt
```

### read data

```python
df = pd.read_excel('cleaned_movies_data.xlsx')
```

### Distribution of BoxOffice Collections

```python
plt.figure(figsize=(10, 6))
plt.hist(df['boxoffice'], bins=30, edgecolor='black', alpha=0.7)
plt.title('Distribution of BoxOffice Collections')
plt.xlabel('BoxOffice Collections')
plt.ylabel('Frequency')
plt.tight_layout()
plt.savefig('boxoffice_distribution.png')
plt.show()
```

### Outliers

```python
Q1 = df['boxoffice'].quantile(0.25)
Q3 = df['boxoffice'].quantile(0.75)
IQR = Q3 - Q1
outliers = df[df['boxoffice'] > Q3 + 1.5 * IQR]
print("Potential outliers:")
print(outliers[['movie_name', 'boxoffice']])
```
