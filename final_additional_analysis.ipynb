```python
import pandas as pd
import matplotlib.pyplot as plt
```

### read data

```python
df = pd.read_excel('cleaned_movies_data.xlsx')
```

### Genre Popularity Across Industries

```python
genre_counts = df.groupby(['industry', 'genre']).size().unstack().fillna(0)
genre_counts.T.plot(kind='bar', stacked=True, figsize=(12, 6), color=plt.cm.Paired.colors)
plt.title("Genre Popularity Across Industries")
plt.xlabel("Genre")
plt.ylabel("Number of Movies")
plt.tight_layout()
plt.savefig("genre_popularity_across_industries.png")
plt.show()
```

### ROI Distribution by Industry

```python
plt.figure(figsize=(10, 6))
df['roi'] = (df['boxoffice'] - df['budget']) / df['budget']

# Use matplotlib to plot the boxplot
industry_groups = df.groupby('industry')['roi'].apply(list)

# Plot a boxplot using matplotlib
plt.boxplot(industry_groups, labels=industry_groups.index, patch_artist=True, 
            boxprops=dict(facecolor='lightblue', color='black'),
            medianprops=dict(color='red'))

plt.title("ROI Distribution by Industry")
plt.xlabel("Industry")
plt.ylabel("ROI")
plt.tight_layout()
plt.savefig("roi_distribution_by_industry.png")
plt.show()
```
