```python
import pandas as pd
import matplotlib.pyplot as plt
```

### cread data from excel

```python
df = pd.read_excel('cleaned_movies_data.xlsx')
```

### calculate roi and profit

```python
df['profit'] = df['boxoffice'] - df['budget']
df['roi'] = df['profit'] / df['budget']
```

```python
print(f'Profit is {df['profit']} and Roi is {df['roi']}')
top_all = df.sort_values('roi', ascending=False).head(3)
```

```python
top_bollywood = df[df['industry'] == 'Bollywood'].sort_values('roi', ascending=False).head(3)
top_tollywood = df[df['industry'] == 'Tollywood'].sort_values('roi', ascending=False).head(3)
```

```python
def plot_top_roi(data, title, filename):
    plt.figure(figsize=(8, 5))
    plt.barh(data['movie_name'], data['roi'], color='skyblue')
    plt.title(title)
    plt.xlabel("ROI")
    plt.tight_layout()
    plt.savefig(f"{filename}.png")  # Save in the current directory
    plt.show()
```

### Graphs


```python
plot_top_roi(top_all, "Top 3 Movies with Highest ROI", "top_roi_all")
plot_top_roi(top_bollywood, "Top 3 Bollywood Movies with Highest ROI", "top_roi_bollywood")
plot_top_roi(top_tollywood, "Top 3 Tollywood Movies with Highest ROI", "top_roi_tollywood")
```
