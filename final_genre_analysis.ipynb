```python
import pandas as pd
import matplotlib.pyplot as plt
```

### read data

```python
df = pd.read_excel('cleaned_movies_data.xlsx')
```

### Average BoxOffice Collection by Genre

```python
avg_by_genre_all = df.groupby('genre')['boxoffice'].mean().sort_values(ascending=False)

plt.figure(figsize=(10, 6))
plt.barh(avg_by_genre_all.index, avg_by_genre_all.values, color='skyblue')
plt.title("Average BoxOffice Collection by Genre (All Industries)")
plt.xlabel("Average BoxOffice")
plt.ylabel("Genre")
plt.tight_layout()
plt.savefig("avg_boxoffice_by_genre_all.png")
plt.show()
```

### Average BoxOffice Collection by Genre (Bollywood)

```python
bolly_df = df[df['industry'] == 'Bollywood']
avg_by_genre_bolly = bolly_df.groupby('genre')['boxoffice'].mean().sort_values(ascending=False)

plt.figure(figsize=(10, 6))
plt.barh(avg_by_genre_bolly.index, avg_by_genre_bolly.values, color='lightgreen')
plt.title("Average BoxOffice Collection by Genre (Bollywood)")
plt.xlabel("Average BoxOffice")
plt.ylabel("Genre")
plt.tight_layout()
plt.savefig("avg_boxoffice_by_genre_bollywood.png")
plt.show()
```
